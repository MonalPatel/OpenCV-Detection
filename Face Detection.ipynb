# Face Detection In Python Using OpenCV

OpenCV contains many pre-trained classifiers for face, eyes, smile etc.There are two pre-trained classifiers:

1.Haar Cascade Classifier

2.LBP Cascade Classifier

### Dependencies


OpenCV should be installed.

Python v3 should be installed.

(Optional) Matplotlib should be installed if you want to see organized results like I have shown in this tutorial. But it is completely optional.

### Importing Required Liberaries 

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import time 
```

```python
def convertToRGB(img):
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

### Code - Haar Cascade Classifier

```python
start = time.time()
#load cascade classifier training file for haarcascade
haar_face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')

#load test iamge
test1 = cv2.imread('Monal.jpg')

#convert the test image to gray image as opencv face detector expects gray images
gray_img = cv2.cvtColor(test1, cv2.COLOR_BGR2GRAY)

#display the gray image using OpenCV
# cv2.imshow('Test Imag', gray_img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

#or if you have matplotlib installed then 
plt.imshow(gray_img, cmap='gray')
elapsed = (time.time() - start)
print("Time taken %s seconds" % (time.time() - start))
```

Now I find the faces in the image with detectMultiScale. If faces are found, this function returns the positions of detected faces as Rect(x,y,w,h).

```python
#let's detect multiscale (some images may be closer to camera than others) images
faces = haar_face_cascade.detectMultiScale(gray_img, scaleFactor=1.1, minNeighbors=5);

#print the number of faces found
print('Faces found: ', len(faces))
```

```python
start = time.time()
#go over list of faces and draw them as rectangles on original colored img
for (x, y, w, h) in faces:
    cv2.rectangle(test1, (x, y), (x+w, y+h), (0, 255, 0), 2)
#convert image to RGB and show image
plt.imshow(convertToRGB(test1))
elapsed = (time.time() - start)
print("Time taken %s seconds" % (time.time() - start))
```
