# Car Detection

In this Notebook we will look at car detection using haar features. We have a haar cascade file trained on cars.The program will detect regions of interest, classify them as cars and show rectangles around them.

### Step:1 Import OpenCV

```python
import cv2

# Will use matplotlib for showing the image in notebook

from matplotlib import pyplot as plt
```

### Step:2 Detecting with cascades

The haar-cascade "cars.xml" was trained using 526 images of cars from the rear (360 x 240 pixels, no scale).You need to download cars.xml file and image of cars.

```python
car_cascade = cv2.CascadeClassifier('cars.xml')
img = cv2.imread('car.jpg', 1)


# Haarcascade classifiersupport support gray sclae image so 
# we need convert colour image to gray scale. 

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Detect cars
cars = car_cascade.detectMultiScale(gray, 1.1, 1)

# Draw border
for (x, y, w, h) in cars:
    cv2.rectangle(img, (x,y), (x+w,y+h), (0,0,255), 2)
   

# Show image
plt.figure(figsize=(10,20))
plt.imshow(img)
```
